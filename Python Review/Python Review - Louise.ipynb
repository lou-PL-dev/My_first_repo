```python
# Creating and accessing lists
fruits = ["apple", "banana", "orange"]
numbers = [1, 2, 3, 4, 5]
 
# Indexing (starts at 0)
first_fruit = fruits[0]  # "apple"
last_fruit = fruits[-1]  # "orange"
 
# Common methods
fruits.append("grape")  # Add to end
fruits.pop()  # Remove last item
fruits.insert(1, "mango")  # Insert at position
```

```python
# Creating dictionaries
student = {
    "name": "Alice",
    "age": 25,
    "course": "AI Bootcamp"
}
 
# Accessing values
name = student["name"]  # "Alice"
age = student.get("age")  # 25 (safer method)
 
# Common methods
student["email"] = "alice@example.com"  # Add/update
del student["age"]  # Remove key
keys = student.keys()  # Get all keys
values = student.values()  # Get all values
```

```python
# Creating sets
numbers = {1, 2, 3, 4, 5}
colors = {"red", "blue", "green"}
 
# Set operations
numbers.add(6)  # Add element
numbers.remove(3)  # Remove element
union = numbers | {7, 8}  # Union
intersection = numbers & {4, 5, 6}  # Intersection
```

```python
# Basic if/else
age = 20
if age >= 18:
    status = "adult"
else:
    status = "minor"
 
# Multiple conditions with elif
score = 85
if score >= 90:
    grade = "A"
elif score >= 80:
    grade = "B"
elif score >= 70:
    grade = "C"
else:
    grade = "F"
```

```python
# Iterating over a list
fruits = ["apple", "banana", "orange"]
for fruit in fruits:
    print(fruit)
 
# Using range
for i in range(5):  # 0, 1, 2, 3, 4
    print(i)
 
# With index
for index, fruit in enumerate(fruits):
    print(f"{index}: {fruit}")
```

```python
# Import pandas (usually aliased as pd)
import pandas as pd
 
# Creating a DataFrame from a dictionary
data = {
    "name": ["Alice", "Bob", "Charlie"],
    "age": [25, 30, 35],
    "score": [85, 90, 88]
}
df = pd.DataFrame(data)
 
# Basic operations
print(df.head())  # Display first few rows
print(df["age"].mean())  # Calculate mean age
print(df[df["score"] > 85])  # Filter rows
 
# Reading from CSV
df_from_file = pd.read_csv("data.csv")
 
# Writing to CSV
df.to_csv("output.csv", index=False)
```

```python
# Import json library
import json
 
# JSON string (like what you get from an API)
json_string = '{"name": "Alice", "age": 25, "course": "AI Bootcamp"}'
 
# Parse JSON string to Python dictionary
data = json.loads(json_string)
print(data["name"])  # "Alice"
 
# Convert Python dictionary to JSON string
student_dict = {
    "name": "Bob",
    "age": 30,
    "grades": [85, 90, 88]
}
json_output = json.dumps(student_dict)
print(json_output)  # '{"name": "Bob", "age": 30, "grades": [85, 90, 88]}'
 
# Reading from a JSON file
with open("data.json", "r") as file:
    data_from_file = json.load(file)
 
# Writing to a JSON file
with open("output.json", "w") as file:
    json.dump(student_dict, file, indent=2)
```
